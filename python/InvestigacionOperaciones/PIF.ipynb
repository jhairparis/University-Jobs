## Conjuntos

## Transporte

Plantas {PL1, PL2} $\rightarrow j$  

Centro de Distribucion {CD1, CD2} $\rightarrow k$  

Clientes {CL1, CL2, CL3, CL4} $\rightarrow m$

## Transbordo 

Provedores {P1, P2, P3} $\rightarrow i$

Plantas {PL1, PL2} $\rightarrow j$  

Insumos = $v\rightarrow$ {I1,I2,I3,I4,I5,I6,I7,I8}

Productos = $l\rightarrow$ {PR1,PR2,PR3,PR4,PR5,PR6,PR7}

$X_{l}$ = Cantidad de productos

$Y_{l,v}$ = (Insumos x Producto) Cantidad de insumo $v$ para hacer producto $l$
